### 4.3 Financial Anlysis

#### 4.3.1 Income Statment

```python
income_sorted = income.sort_values("Date")

income_sorted["Revenue_Growth"] = income_sorted["Total_Revenue"].pct_change() * 100

income_sorted["Gross_Margin"] = (income_sorted["Gross_Profit"] / income_sorted["Total_Revenue"]) * 100

income_sorted["EBIT_Margin"] = (income_sorted["EBIT"] / income_sorted["Total_Revenue"]) * 100

income_sorted["Net_Margin"] = (income_sorted["Net_Income"] / income_sorted["Total_Revenue"]) * 100

income_sorted["RD_Pct"] = (income_sorted["Research_And_Development"] / income_sorted["Total_Revenue"]) * 100

cols = ["Date", "Total_Revenue", "Revenue_Growth", "Gross_Margin", 
        "EBIT_Margin", "Net_Margin", "RD_Pct"]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Revenue Growth
axes[0,0].bar(income_sorted["Date"].dt.year, income_sorted["Revenue_Growth"], 
              color=["red" if x < 0 else "steelblue" for x in income_sorted["Revenue_Growth"]])
for i, (date, val) in enumerate(zip(income_sorted["Date"].dt.year, income_sorted["Revenue_Growth"])):
    if not pd.isna(val):
        axes[0,0].text(date, val + 0.2, f"{val:.1f}%", ha="center", fontsize=10, fontweight="bold")
axes[0,0].set_title("Revenue Growth %", fontsize=12, fontweight="bold")
axes[0,0].set_xlabel("Year")

# Gross Margin
axes[0,1].bar(income_sorted["Date"].dt.year, income_sorted["Gross_Margin"], color="green", alpha=0.7)
for i, (date, val) in enumerate(zip(income_sorted["Date"].dt.year, income_sorted["Gross_Margin"])):
    axes[0,1].text(date, val - 2, f"{val:.1f}%", ha="center", fontsize=10, fontweight="bold", color="white")
axes[0,1].set_title("Gross Margin %", fontsize=12, fontweight="bold")
axes[0,1].set_ylim(60, 72)

# Net Profit Margin
axes[1,0].bar(income_sorted["Date"].dt.year, income_sorted["Net_Margin"], color="orange", alpha=0.8)
for i, (date, val) in enumerate(zip(income_sorted["Date"].dt.year, income_sorted["Net_Margin"])):
    axes[1,0].text(date, val - 1.5, f"{val:.1f}%", ha="center", fontsize=10, fontweight="bold", color="white")
axes[1,0].set_title("Net Profit Margin %", fontsize=12, fontweight="bold")
axes[1,0].set_ylim(30, 40)

# R&D % of Revenue
axes[1,1].bar(income_sorted["Date"].dt.year, income_sorted["RD_Pct"], color="red", alpha=0.7)
for i, (date, val) in enumerate(zip(income_sorted["Date"].dt.year, income_sorted["RD_Pct"])):
    axes[1,1].text(date, val - 0.5, f"{val:.1f}%", ha="center", fontsize=10, fontweight="bold", color="white")
axes[1,1].set_title("R&D as % of Revenue", fontsize=12, fontweight="bold")

plt.suptitle("4.3 Income Statement Analysis", fontsize=16, fontweight="bold", y=1.01)
plt.tight_layout()
plt.savefig("charts/income_analysis.png", dpi=150, bbox_inches="tight")
plt.show()
```

**Revenue Growth:**

- 2023→2024: **+15.7%** — strong acceleration driven by Azure and AI
- 2024→2025: **+14.9%** — continued double-digit growth

**Gross Margin (~68-70%):**
- Consistently high across all years — reflects Microsoft's software and 
  cloud pricing power
- Peak at **69.8% in 2024** — most efficient year

**Net Profit Margin (~34-37%):**
- Exceptional profitability for a company of this scale
- Dip to **34.1% in 2023** due to increased AI investment costs
- Recovered to **36.1% in 2025**

**R&D as % of Revenue (~11-13%):**
- Heavy and consistent investment in research and development
- Declining % as revenue grows faster than R&D spend
- Reflects Microsoft's shift from pure R&D to commercializing AI

**Business Insight:**
Microsoft demonstrates best-in-class profitability metrics. The combination 
of high gross margins and strong net margins reflects the scalability of 
its cloud and software business model. The acceleration in revenue growth

```python
balance = pd.read_csv("data/MSFT_balance.csv")
balance["Date"] = pd.to_datetime(balance["Date"])
balance = balance[balance["Date"].dt.year != 2021]
balance_sorted = balance.sort_values("Date")
print(balance_sorted.shape)
```

```python
financial = income_sorted.merge(balance_sorted, on="Date", how="inner")
print(financial.shape)
print(financial["Date"].values)
```

```python
# Profitability Ratios
financial["Gross_Margin"] = financial["Gross_Profit"] / financial["Total_Revenue"] * 100
financial["EBIT_Margin"] = financial["EBIT"] / financial["Total_Revenue"] * 100
financial["Net_Margin"] = financial["Net_Income"] / financial["Total_Revenue"] * 100
financial["Revenue_Growth"] = financial["Total_Revenue"].pct_change() * 100
financial["RD_Pct"] = financial["Research_And_Development"] / financial["Total_Revenue"] * 100

# Liquidity Ratios
financial["Current_Ratio"] = financial["Current_Assets"] / financial["Current_Liabilities"]
financial["Quick_Ratio"] = (financial["Current_Assets"] - financial["Inventory"]) / financial["Current_Liabilities"]
financial["Cash_Ratio"] = financial["Cash_And_Cash_Equivalents"] / financial["Current_Liabilities"]

# Leverage Ratios
financial["Debt_To_Equity"] = financial["Total_Debt"] / financial["Stockholders_Equity"]
financial["Debt_To_Assets"] = financial["Total_Debt"] / financial["Total_Assets"]

# Efficiency Ratios
financial["Asset_Turnover"] = financial["Total_Revenue"] / financial["Total_Assets"]
financial["ROE"] = financial["Net_Income"] / financial["Stockholders_Equity"] * 100
financial["ROA"] = financial["Net_Income"] / financial["Total_Assets"] * 100

# Print summary
cols = ["Date", "Gross_Margin", "EBIT_Margin", "Net_Margin",
        "Current_Ratio", "Debt_To_Equity", "ROE", "ROA"]
print(financial[cols].to_string())
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

years = financial["Date"].dt.year

# Liquidity Ratios
axes[0,0].bar(years, financial["Current_Ratio"], color="steelblue")
axes[0,0].set_title("Current Ratio")
axes[0,0].axhline(y=1, color="red", linestyle="--", label="Min Safe Level")
for y, v in zip(years, financial["Current_Ratio"]):
    axes[0,0].text(y, v-0.1, f"{v:.2f}", ha="center", color="white", fontweight="bold")

axes[0,1].bar(years, financial["Quick_Ratio"], color="orange")
axes[0,1].set_title("Quick Ratio")
for y, v in zip(years, financial["Quick_Ratio"]):
    axes[0,1].text(y, v-0.1, f"{v:.2f}", ha="center", color="white", fontweight="bold")

axes[0,2].bar(years, financial["Cash_Ratio"], color="green")
axes[0,2].set_title("Cash Ratio")
for y, v in zip(years, financial["Cash_Ratio"]):
    axes[0,2].text(y, v-0.05, f"{v:.2f}", ha="center", color="white", fontweight="bold")

# Leverage Ratios
axes[1,0].bar(years, financial["Debt_To_Equity"], color="red")
axes[1,0].set_title("Debt to Equity")
for y, v in zip(years, financial["Debt_To_Equity"]):
    axes[1,0].text(y, v-0.02, f"{v:.2f}", ha="center", color="white", fontweight="bold")

axes[1,1].bar(years, financial["Debt_To_Assets"], color="purple")
axes[1,1].set_title("Debt to Assets")
for y, v in zip(years, financial["Debt_To_Assets"]):
    axes[1,1].text(y, v-0.01, f"{v:.2f}", ha="center", color="white", fontweight="bold")

axes[1,2].bar(years, financial["ROE"], color="teal")
axes[1,2].set_title("Return on Equity %")
for y, v in zip(years, financial["ROE"]):
    axes[1,2].text(y, v-2, f"{v:.1f}%", ha="center", color="white", fontweight="bold")

plt.suptitle("4.4 Balance Sheet Analysis", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.savefig("charts/balance_sheet_ratios.png", dpi=150, bbox_inches="tight")
plt.show()
```

**Liquidity Ratios:**
- **Current Ratio** declined from 1.78 → 1.35 — still above 1.0 (safe) 
  but tightening as Microsoft deploys cash into AI investments
- **Quick Ratio** mirrors current ratio — Microsoft carries minimal 
  inventory as a software/cloud company
- **Cash Ratio** fluctuates — peaked at 0.33 in 2023 (OpenAI investment 
  year) then declined as cash was deployed

**Leverage Ratios:**
- **Debt to Equity** dropped dramatically from 0.37 → 0.18 — Microsoft 
  is deleveraging while growing, very conservative financial structure
- **Debt to Assets** also declining (0.17 → 0.10) — assets growing 
  faster than debt

**Return on Equity:**
- ROE declining from 43.7% → 29.6% — not a sign of weakness
- Equity base is growing faster than profits as Microsoft retains 
  earnings and issues stock for acquisitions
- Still exceptionally high ROE by any industry standard

**Business Insight:**
Microsoft maintains a fortress balance sheet — low debt, high liquidity, 
and declining leverage while aggressively investing in AI infrastructure. 
This financial strength gives analysts confidence to maintain Bullish 
recommendations even at premium valuations.

```python
cashflow_sorted = cashflow.sort_values("Date")
cashflow_sorted = cashflow_sorted[cashflow_sorted["Date"].dt.year != 2021]

# Merge with financial
financial2 = financial.merge(cashflow_sorted, on="Date", how="inner")

# Cash Flow Ratios
financial2["OCF_Ratio"] = financial2["Operating_Cash_Flow"] / financial2["Current_Liabilities"]
financial2["FCF"] = financial2["Free_Cash_Flow"]
financial2["Cash_Flow_Margin"] = financial2["Operating_Cash_Flow"] / financial2["Total_Revenue"]
financial2["Cash_ROA"] = financial2["Operating_Cash_Flow"] / financial2["Total_Assets"]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

years2 = financial2["Date"].dt.year

# OCF Ratio
axes[0,0].bar(years2, financial2["OCF_Ratio"], color="steelblue")
axes[0,0].set_title("Operating Cash Flow Ratio")
for y, v in zip(years2, financial2["OCF_Ratio"]):
    axes[0,0].text(y, v-0.05, f"{v:.2f}", ha="center", color="white", fontweight="bold")

# Free Cash Flow
axes[0,1].bar(years2, financial2["FCF"]/1e9, color="green")
axes[0,1].set_title("Free Cash Flow (Billions USD)")
for y, v in zip(years2, financial2["FCF"]/1e9):
    axes[0,1].text(y, v-2, f"${v:.1f}B", ha="center", color="white", fontweight="bold")

# Cash Flow Margin
axes[1,0].bar(years2, financial2["Cash_Flow_Margin"]*100, color="orange")
axes[1,0].set_title("Cash Flow Margin %")
for y, v in zip(years2, financial2["Cash_Flow_Margin"]*100):
    axes[1,0].text(y, v-2, f"{v:.1f}%", ha="center", color="white", fontweight="bold")

# Cash ROA
axes[1,1].bar(years2, financial2["Cash_ROA"]*100, color="purple")
axes[1,1].set_title("Cash Return on Assets %")
for y, v in zip(years2, financial2["Cash_ROA"]*100):
    axes[1,1].text(y, v-1, f"{v:.1f}%", ha="center", color="white", fontweight="bold")

plt.suptitle("4.5 Cash Flow Analysis", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.savefig("charts/cashflow_ratios.png", dpi=150, bbox_inches="tight")
plt.show()
```

**Operating Cash Flow Ratio (0.84 - 0.95):**
- Consistently below 1.0 but healthy for a large tech company
- 2023 dip to 0.84 — increased working capital needs during AI expansion
- Recovered to 0.95 in 2024 — improving cash generation efficiency

**Free Cash Flow:**
- **$65.1B (2022)** → **$59.5B (2023)** → **$74.1B (2024)**
- Dip in 2023 driven by heavy capital expenditure on AI infrastructure
- Strong recovery in 2024 — $74.1B is exceptional free cash flow
- Microsoft is one of the highest FCF generators in the world

**Cash Flow Margin (41-48%):**
- Every dollar of revenue generates ~$0.45 in operating cash
- Improving trend — peaked at 48.4% in 2024
- Reflects the scalability of cloud and software business model

**Cash Return on Assets (21-24%):**
- Consistently strong — Microsoft extracts significant cash from its 
  asset base
- Slight decline in 2023 as asset base grew with AI investments

**Business Insight:**
Microsoft's free cash flow of $74.1B in 2024 gives it extraordinary 
financial flexibility — funding AI investments, dividends, buybacks, 
and acquisitions simultaneously without stress. This cash generation 
capacity is a key reason analysts remain persistently Bullish.
