* there are 468 nulls in ***FROMGrade*** and they are expected and meaningful not a data error , because when analyst covers stock for the first time there is no previous grade

```python
upgrades_downgrades["FromGrade"] = upgrades_downgrades["FromGrade"].fillna("New Coverage")
```

```python
print(upgrades_downgrades["ToGrade"].value_counts())
print(upgrades_downgrades.shape)
```

```python
upgrades_downgrades = upgrades_downgrades.dropna(subset=["ToGrade"])
upgrades_downgrades["GradeDate"] = pd.to_datetime(upgrades_downgrades["GradeDate"])
```

```python
income["Date"] = pd.to_datetime(income["Date"])
balance["Date"] = pd.to_datetime(balance["Date"])
cashflow["Date"] = pd.to_datetime(cashflow["Date"])
```

```python
income = income.dropna()
balance = balance.dropna()
cashflow = cashflow.dropna()
```

```python
price["Date"] = pd.to_datetime(price["Date"])
price = price.sort_values("Date")
```

```python
bullish = ["Buy", "Outperform", "Overweight", "Strong Buy", 
           "Sector Outperform", "Market Outperform", "Long-Term Buy"]

bearish = ["Sell", "Underperform", "Underweight"]
```

```python
def map_recommendation (grade):
    if grade in bullish:
        return 2
    elif grade in bearish :
        return 0
    else:
        return 1

upgrades_downgrades["target"] = upgrades_downgrades["ToGrade"].apply(map_recommendation)
print(upgrades_downgrades["target"].value_counts())
```
